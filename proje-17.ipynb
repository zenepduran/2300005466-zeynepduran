```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QLabel, QLineEdit, QPushButton, QVBoxLayout, QWidget


class EgitimPlatformu(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("Eğitim Platformu")
        self.setGeometry(100, 100, 400, 300)  # Pencere boyutunu ve konumunu ayarlayın

        self.initUI()

    def initUI(self):
        self.central_widget = QWidget()  # Merkezi bir widget oluşturun
        self.setCentralWidget(self.central_widget)

        self.layout = QVBoxLayout()  # Dikey bir düzen oluşturun

        self.label_ad = QLabel("Öğrenci Adı ve Soyadı:")
        self.edit_ad = QLineEdit()
        self.layout.addWidget(self.label_ad)
        self.layout.addWidget(self.edit_ad)

        self.label_ders = QLabel("Dersler:")
        self.edit_ders = QLineEdit()
        self.layout.addWidget(self.label_ders)
        self.layout.addWidget(self.edit_ders)

        self.label_kaynak = QLabel("Ders Kaynakları:")
        self.edit_kaynak = QLineEdit()
        self.layout.addWidget(self.label_kaynak)
        self.layout.addWidget(self.edit_kaynak)

        self.label_ogretmen = QLabel("Ders Öğretmeni:")
        self.edit_ogretmen = QLineEdit()
        self.layout.addWidget(self.label_ogretmen)
        self.layout.addWidget(self.edit_ogretmen)

        self.button_kaydet = QPushButton("Kaydet")
        self.button_kaydet.clicked.connect(self.kaydet)
        self.layout.addWidget(self.button_kaydet)

        # Bilgileri göstermek için QLabel'lar
        self.label_goster_ad = QLabel("")
        self.layout.addWidget(self.label_goster_ad)

        self.label_goster_ders = QLabel("")
        self.layout.addWidget(self.label_goster_ders)

        self.label_goster_kaynak = QLabel("")
        self.layout.addWidget(self.label_goster_kaynak)

        self.label_goster_ogretmen = QLabel("")
        self.layout.addWidget(self.label_goster_ogretmen)

        self.central_widget.setLayout(self.layout)

    def kaydet(self):
        ad_soyad = self.edit_ad.text()
        dersler = self.edit_ders.text()
        kaynaklar = self.edit_kaynak.text()
        ogretmen = self.edit_ogretmen.text()

        # QLabel'lara kaydedilen bilgileri göster
        self.label_goster_ad.setText(f"Öğrenci Adı ve Soyadı: {ad_soyad}")
        self.label_goster_ders.setText(f"Dersler: {dersler}")
        self.label_goster_kaynak.setText(f"Ders Kaynakları: {kaynaklar}")
        self.label_goster_ogretmen.setText(f"Ders Öğretmeni: {ogretmen}")


if __name__ == "__main__":
    app = QApplication(sys.argv)
    egitim_platformu = EgitimPlatformu()
    egitim_platformu.show()
    sys.exit(app.exec_())
```
